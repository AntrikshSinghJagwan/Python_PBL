```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv("cleaned_dataset.csv")
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Accident_severity'
)

plt.title("Accident Severity Distribution")
plt.show()
```

Dataset is imbalanced

```python
sns.catplot(
    data=df,
    x='Age_band_of_driver',
    hue='Accident_severity',
    kind='count',
    height=6,
    aspect=2
)

plt.xticks(rotation=45)
plt.show()
```

```python
df["Age_band_of_driver"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(8,8)
)

plt.title("Driver Age Distribution")
plt.ylabel("")
plt.show()
```

```python
pd.crosstab(
    df["Driving_experience"],
    df["Accident_severity"]
).plot(
    kind="bar",
    stacked=True,
    figsize=(12,6)
)

plt.title("Driving Experience vs Severity")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.catplot(
    data=df,
    x="Weather_conditions",
    hue="Accident_severity",
    kind="count",
    height=6,
    aspect=2
)

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Accident_severity",
    y="Number_of_casualties"
)

plt.title("Casualties by Severity")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.violinplot(
    data=df,
    x="Accident_severity",
    y="Number_of_vehicles_involved"
)

plt.title("Vehicles Involved by Severity")
plt.show()
```

```python
plt.figure(figsize=(12,8))

df["Cause_of_accident"].value_counts().plot(
    kind="barh"
)

plt.title("Most Common Causes of Accidents")
plt.show()
```

```python
cross = pd.crosstab(
    df["Light_conditions"],
    df["Accident_severity"]
)

plt.figure(figsize=(8,5))

sns.heatmap(
    cross,
    annot=True,
    cmap="YlOrRd"
)

plt.title("Light Conditions vs Severity")
plt.show()
```

```python
df["Road_surface_conditions"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    wedgeprops=dict(width=0.4),
    figsize=(8,8)
)

plt.title("Road Surface Conditions")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.show()
```

Shows almost zero correlation between no. of vehicles involved and no. of casualties
