```python
import joblib
import pandas as pd

# Load model and preprocessing files

'''
Available models:
"svm_rbf_model.pkl"
"svm_linear_model.pkl"
"svm_poly_model.pkl"
"svm_sigmoid_model.pkl"
"logistic_model.pkl"
'''
model = joblib.load("logistic_model.pkl")   # change model here
scaler = joblib.load("scaler.pkl")
selected_features = joblib.load("selected_features.pkl")
target_encoder = joblib.load("target_encoder.pkl")

''' # severe injury case
user_data = {
    "Age_band_of_driver": "under 18",
    "Sex_of_driver": "male",
    "Educational_level": "Low Education",
    "Vehicle_driver_relation": "employee",
    "Driving_experience": "New Driver",
    "Type_of_vehicle": "Heavy Vehicle",
    "Owner_of_vehicle": "organization",
    "Area_accident_occured": "Rural",
    "Lanes_or_Medians": "undivided two way",
    "Road_allignment": "Steep/Sloped Terrain",
    "Road_surface_type": "gravel roads",
    "Type_of_collision": "Human Collision",
    "Vehicle_movement": "Vehicle Exit/Loss of Control",
    "Cause_of_accident": "Speed Violation",
    "Light_conditions": "darkness - no lighting",
    "Weather_conditions": "Rain",
    "Road_surface_conditions": "wet or damp",
    "Number_of_vehicles_involved": 6,
    "Number_of_casualties": 8
}
'''

''' # severe injury case
user_data = {
    "Age_band_of_driver": "31-50",
    "Sex_of_driver": "male",
    "Educational_level": "Basic Education",
    "Vehicle_driver_relation": "employee",
    "Driving_experience": "Intermediate",
    "Type_of_vehicle": "Heavy Vehicle",
    "Owner_of_vehicle": "organization",
    "Area_accident_occured": "Commercial/Workplace",
    "Lanes_or_Medians": "undivided two way",
    "Road_allignment": "Curved Road",
    "Road_surface_type": "asphalt roads with some distress",
    "Type_of_collision": "Vehicle Collision",
    "Vehicle_movement": "Vehicle Exit/Loss of Control",
    "Cause_of_accident": "Vehicle Loss of Control",
    "Light_conditions": "darkness - no lighting",
    "Weather_conditions": "Rain",
    "Road_surface_conditions": "wet or damp",
    "Number_of_vehicles_involved": 3,
    "Number_of_casualties": 7
}
'''
# slight injury case
user_data = {
    "Age_band_of_driver": "31-50",
    "Sex_of_driver": "male",
    "Educational_level": "Higher Education",
    "Vehicle_driver_relation": "owner",
    "Driving_experience": "Veteran",
    "Type_of_vehicle": "Private Vehicle",
    "Owner_of_vehicle": "owner",
    "Area_accident_occured": "Urban/Residential",
    "Lanes_or_Medians": "double carriageway (median)",
    "Road_allignment": "Straight Road",
    "Road_surface_type": "asphalt roads",
    "Type_of_collision": "Object Collision",
    "Vehicle_movement": "Moving Forward",
    "Cause_of_accident": "Other",
    "Light_conditions": "daylight",
    "Weather_conditions": "Clear",
    "Road_surface_conditions": "dry",
    "Number_of_vehicles_involved": 1,
    "Number_of_casualties": 1
}

# Convert to DataFrame
input_df = pd.DataFrame([user_data])

# One-hot encode
input_df = pd.get_dummies(input_df)

# Add missing selected columns
for col in selected_features:
    if col not in input_df.columns:
        input_df[col] = 0

# Keep exact feature order
input_df = input_df[selected_features]

# Scale
input_scaled = scaler.transform(input_df)

# Predict
prediction = model.predict(input_scaled)

# Decode prediction
predicted_label = target_encoder.inverse_transform(prediction)

print("Predicted Accident Severity:")
print(predicted_label[0])

# Show probabilities if available
if hasattr(model, "predict_proba"):
    probabilities = model.predict_proba(input_scaled)

    print("\nPrediction Confidence:")
    for label, prob in zip(target_encoder.classes_, probabilities[0]):
        print(label, ":", round(prob * 100, 2), "%")
```

This project analyzed road accident severity prediction using Support Vector Machine with different kernel functions (Linear, RBF, Polynomial, and Sigmoid) and Logistic Regression. Data preprocessing included cleaning, category reduction, feature selection, handling class imbalance using SMOTE, and feature scaling to improve model performance.

Among all the models tested, the Polynomial Kernel SVM performed the best in identifying severe accidents. Although the RBF kernel achieved higher overall accuracy, it showed poor performance in detecting severe injury cases due to class imbalance. The Polynomial SVM achieved the highest recall and F1-score for the severe injury class and the best macro-average F1-score, making it the most balanced and reliable model for this problem.

The implementation results also showed that Polynomial SVM correctly classified high-risk accident scenarios as severe injuries, demonstrating better real-world prediction capability. Therefore, Polynomial SVM was selected as the final model for accident severity prediction because it provides a better balance between accuracy and minority-class detection, making it more practical for identifying critical accident cases.
